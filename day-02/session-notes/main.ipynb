## 1. Control Flow: If, Elif, and Else

Control flow allows your program to make decisions based on certain conditions. 

- **`if` statement**: Executes a block of code if its condition is true.
- **`elif` (else if)**: Checks another condition if the previous ones were false.
- **`else`**: Executes if none of the above conditions are true.

### Example:

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: F (Fail)")
```

## 2. While Loops and Repetition

A `while` loop repeats a block of code as long as a specified condition remains true. It is useful when you don't know exactly how many iterations you need.

- **`break` statement**: Used to exit the loop immediately, even if the condition is still true.

### Example: Multiplication Table

```python
num = int(input("Enter a number for the multiplication table: "))
count = 1

while count <= 10:
    print(f"{num} x {count} = {num * count}")
    count += 1
```

## 3. Introduction to Lists

A **List** is an ordered collection of items. It is changeable (mutable) and allows duplicate values.

- **Indexing**: Access items by position (starts at 0).
- **`append()`**: Adds an item to the end.
- **`remove()`**: Removes a specific item by value.
- **`len()`**: Returns the number of items in the list.

### Example:

```python
names = ["Ram", "Sam", "John"]
names.append("Arun")
print("List after append:", names)

names.remove("Sam")
print("List after remove:", names)

print("Last element:", names[-1])
print("Total elements:", len(names))
```

## 4. Dictionaries

A **Dictionary** stores data in **Key-Value pairs**. It is fast for data retrieval and behaves like a real-world dictionary.

- **Syntax**: `{ "key": "value" }` 
- **`.items()`**: Returns both keys and values for looping.

### Example:

```python
student = {
    "name": "Arun",
    "age": 20,
    "grade": "A"
}

print(f"Student Name: {student['name']}")

# Modifying and adding data
student["age"] = 21
student["city"] = "Chennai"

for key, value in student.items():
    print(f"{key.capitalize()}: {value}")
```

## 5. Mini Project: Student Management System

Combining everything (Lists, Dictionaries, Loops, and Input), here is a simple menu-based program to manage student records.

### Project Outline:

```python
students = []

while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Exit")
    
    choice = input("Enter your choice (1-3): ")
    
    if choice == "1":
        name = input("Enter student name: ")
        age = input("Enter student age: ")
        students.append({"name": name, "age": age})
        print("Student added!")
    elif choice == "2":
        if not students:
            print("No records found.")
        else:
            for s in students:
                print(f"Name: {s['name']}, Age: {s['age']}")
    elif choice == "3":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid input, please try again.")
```
