```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
generator = pipeline("text-generation", model="gpt2")

result = generator(
    "Artificial Intelligence will change the world because",
    max_length=50,
    num_return_sequences=1
)

print(result[0]["generated_text"])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Machine learning is amazing"

tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

```python
text = "Natural Language Processing is interesting"

encoded = tokenizer.encode(text)
decoded = tokenizer.decode(encoded)

print("Encoded:", encoded)
print("Decoded:", decoded)
```

```python
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

input_text = "AI in healthcare will"

inputs = tokenizer(input_text, return_tensors="pt")

outputs = model.generate(
    **inputs,
    max_length=40,
    num_return_sequences=1
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")


```

```python
text = """
Artificial Intelligence is transforming industries by enabling machines
to learn from data and perform tasks that usually require human intelligence.
AI is used in healthcare, finance, transportation, and education.
Machine learning algorithms help analyze large datasets and make predictions.
Deep learning models can recognize images, translate languages, and even generate text.
"""

summary = summarizer(
    text,
    max_length=60,
    min_length=20,
    do_sample=False
)

print(summary[0]["summary_text"])
```
