```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

text = "I love learning NLP. It is amazing!"

# Sentence Tokenization
sentences = sent_tokenize(text)
print("Sentences:", sentences)

# Word Tokenization
words = word_tokenize(text)
print("Words:", words)
```

```python
from nltk.corpus import stopwords

text = "I love learning NLP and it is very interesting"

words = word_tokenize(text)

stop_words = set(stopwords.words("english"))

filtered_words = []

for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)

print("Original:", words)
print("After removing stopwords:", filtered_words)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "cars", "better"]

for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

text = "This phone is amazing and the battery life is great"

score = sia.polarity_scores(text)

print(score)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

text = """
The new smartphone launched today has amazing features.
Many users love the battery life and camera quality.
"""

# Tokenization
words = word_tokenize(text)

# Stopword removal
stop_words = set(stopwords.words("english"))

filtered = []

for w in words:
    if w.lower() not in stop_words:
        filtered.append(w)

# Lemmatization
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered]

print("Cleaned Words:", lemmatized_words)

# Sentiment
sia = SentimentIntensityAnalyzer()

score = sia.polarity_scores(text)

print("Sentiment Score:", score)
```
