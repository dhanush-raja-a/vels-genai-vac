### 1. Basic Function Syntax
A function is defined using the `def` keyword, followed by the function name and parentheses `()`.

```python
def say_hello():
    print("Hello from inside the function!")

# Calling the function
say_hello()
```

### 2. Parameters and Arguments
Parameters are variables listed in the function definition, while arguments are the values sent to the function when it is called.

```python
def greet(name):
    return f"Hello, {name}!"

message = greet("Dhanush")
print(message)
```

### 3. Default Arguments
You can provide a default value to a parameter if no argument is passed.

```python
def power(base, exponent=2):
    return base ** exponent

print(power(4))    # Uses default exponent=2 -> 16
print(power(4, 3)) # Uses provided exponent=3 -> 64
```

### 4. Variable-length Arguments
- `*args`: Allows passing a variable number of positional arguments (collected as a tuple).
- `**kwargs`: Allows passing a variable number of keyword arguments (collected as a dictionary).

```python
def sum_all(*numbers):
    return sum(numbers)

print(f"Sum: {sum_all(1, 2, 3, 4, 5)}")

def print_details(**details):
    for key, value in details.items():
        print(f"{key}: {value}")

print_details(name="Dhanush", role="Developer", city="Chennai")
```

### 5. Scope (Global vs Local)
Variables created inside a function are local to that function.

```python
x = 10 # Global variable

def my_func():
    x = 5 # Local variable
    print(f"Inside function: {x}")

my_func()
print(f"Outside function: {x}")
```

### 6. Lambda Functions
Small anonymous functions defined with the `lambda` keyword.

```python
multiply = lambda a, b : a * b
print(f"10 * 5 = {multiply(10, 5)}")
```

### 7. Docstrings
Used for documenting what a function does. Accessible via `help(function_name)`.

```python
def complex_operation(a, b):
    """
    Performs a complex mathematical operation.
    Args:
        a: First number
        b: Second number
    Returns:
        The result of (a + b) multiplied by (a - b)
    """
    return (a + b) * (a - b)

print(complex_operation(10, 5))
print(complex_operation.__doc__)
```
