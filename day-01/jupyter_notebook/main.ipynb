# 🧵 The Ultimate Guide to String Manipulation in Python
This notebook covers **every essential and advanced technique** for handling strings in Python. 

### Why Strings?
Strings are one of the most common data types. Whether you're working with web data, user input, or AI models, you'll constantly need to clean, format, and search through text.

## 1. Basics: Creation, Concatenation, and Length
- **Concatenation (`+`)**: Combining two strings.
- **Repetition (`*`)**: Repeating a string multiple times.
- **Length (`len()`)**: Counting the total number of characters (including spaces).

```python
s1 = "Hello"
s2 = "World"

print("Concatenation:", s1 + " " + s2)
print("Repetition:", "Ha! " * 3)
print("Length:", len(s1 + " " + s2))
```

## 2. Indexing and Slicing (Advanced)
- **Indexing**: Strings are 0-indexed. `[0]` is the first character, `[-1]` is the last.
- **Slicing**: `[start:stop:step]` gets a range. 
- **Step**: `[::2]` skips every other character. `[::-1]` reverses the string.

```python
text = "PythonScripting"

print("Full String:", text)
print("Index [0]:", text[0])
print("Index [-1]:", text[-1])
print("Slice [0:6]:", text[0:6])         # Up to (but not including) index 6
print("Substring [6:]:", text[6:])       # From index 6 to the end
print("Skip 2 [::2]:", text[::2])        # Pthncrpig
print("Reversed:", text[::-1])           # gnitpircSnohtyP
```

## 3. Case Transformation
Methods to change how letters appear.

```python
msg = "i love PYTHON programming"

print("Upper:", msg.upper())
print("Lower:", msg.lower())
print("Swapcase:", msg.swapcase())       # Swaps lower to upper and vice versa
print("Title:", msg.title())             # Capitalizes every word
print("Capitalize:", msg.capitalize())   # Capitalizes only the first character
```

## 4. String Property Checks (Validation)
Use these to validate user input (e.g., checking if a password is alphanumeric).

```python
s = "Python3"

print(f"'{s}' is Alpha?", s.isalpha())
print(f"'{s}' is Digit?", s.isdigit())
print(f"'{s}' is Alphanumeric?", s.isalnum())
print(f"'{s}' is Lowercase?", s.islower())
print("Is ' ' space?", " ".isspace())
```

## 5. Cleaning: Stripping and Replacing
- **Strip**: Removes whitespace from ends.
- **Replace**: Swaps a substring for another.

```python
email = "   user@example.com   "
print(f"Cleaned Email: '{email.strip()}'")

sentence = "I like apples. Apples are healthy."
print("Replaced:", sentence.replace("apples", "oranges").replace("Apples", "Oranges"))
```

## 6. Searching and Finding
- **Find**: Returns index of first match (or -1 if not found).
- **Count**: Counts occurrences.
- **Start/Endswith**: Checks start or end of string.

```python
bio = "He is a data scientist and he likes Python."

print("Index of 'data':", bio.find("data"))
print("Count 'he' (case-sensitive):", bio.count("he"))
print("Starts with 'He'?:", bio.startswith("He"))
print("Ends with 'Python.'?:", bio.endswith("Python."))
```

## 7. Splitting, Joining, and Partitioning
- **Split**: String -> List.
- **Join**: List -> String.
- **Partition**: Splits string into a tuple of 3 (start, separator, end).

```python
tags = "python,ai,ml,ds"
tag_list = tags.split(",")
print("Split List:", tag_list)

new_tags = " | ".join(tag_list)
print("Joined String:", new_tags)

url = "www.google.com"
print("Partition:", url.partition("."))
```

## 8. Alignment and Padding
Useful for creating neat text-based tables or menus.

```python
title = "MENU"

print(title.center(20, "*"))
print("Salad".ljust(15, ".") + "$10")
print("Pizza".ljust(15, ".") + "$15")

number = "42"
print("Zero padded:", number.zfill(5))  # 00042
```

## 9. Modern String Formatting
**f-strings** (Introduced in Python 3.6) are the fastest and most readable.

```python
name = "Dhanush"
score = 98.567

# f-string basics
print(f"Student: {name}")

# f-string with math and formatting
print(f"Score: {score:.1f}%") # Rounds to 1 decimal place
```

## 10. Escape Characters and Raw Strings
- `\n`: New line.
- `\t`: Tab.
- **Raw Strings (`r""`)**: Prevents backslashes from being treated as escape characters (essential for file paths).

```python
print("Line 1\nLine 2\tTabbed")

# Without 'r', \t would be a tab
path = r"C:\users\tasks\notebook.ipynb"
print("File Path:", path)
```
